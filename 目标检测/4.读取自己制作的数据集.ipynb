```python
import torchvision
from matplotlib import pyplot as plt
from PIL import ImageDraw
```

```python
coco_dataset = torchvision.datasets.CocoDetection(root=r'H:\数据集\coco1-4\ImageSet',
                                                annFile=r'H:\数据集\coco1-4\annotations\instances_default.json')
```

```python
print(coco_dataset[0])
```

```python
image,info = coco_dataset[2]
# plt.imshow(image)
draw = ImageDraw.Draw(image)
for annotation in info:
    x_min,y_min,width,height = annotation['bbox']
    # print(x_min,y_min,width,height)
    draw.rectangle([x_min,y_min,x_min+width,y_min+height],outline='red')

plt.imshow(image)
```
