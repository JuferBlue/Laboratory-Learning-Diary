```python
from PIL import Image

# 打开图片
image_path = 'img.png'  # 替换为你的图片路径
image = Image.open(image_path)

# 查看图片类型
print(f"图片格式: {image.format}")

# 获取图片尺寸
width, height = image.size
print(f"图片大小: {width}x{height}")

# 遍历每个像素点
for y in range(height):
    for x in range(width):
        # 获取像素颜色值
        pixel_color = image.getpixel((x, y))
        print(f"坐标 ({x},{y}) 的像素颜色: {pixel_color}")
```

```python
from PIL import Image
import numpy as np

# 加载图片
image_path = 'img.png'  # 替换为你的图片路径
image = Image.open(image_path)

# 将图片转换为 NumPy 数组
image_array = np.array(image)

# 显示数组形状和数据类型（可选）
print("Image shape:", image_array.shape)
print("Data type:", image_array.dtype)

# 保存 NumPy 数组到 .npy 文件
output_file = 'saved_image.npy'
np.save(output_file, image_array)

print(f"Image has been saved to {output_file}")
```

```python
import os
path = './data/NEU-Seg/ImageSets/Segmentation/test.txt'
names = []
with open(path, 'r') as f:
    lines = f.readlines()
    for line in lines:
        name = line.strip()
        names.append(name)

for i in names:
    print(i)
img_path = './data/NEU-Seg/SegmentationClass'
save_path = './test_ground_truths'
for i in range(len(names)):
    img_name = names[i] + '.png'
    image = Image.open(img_path + '/' + img_name)
    # 将图片转换为numpy数组
    image_array = np.array(image)
    # 保存为npy文件
    np.save(os.path.join(save_path, 'ground_truth_'+names[i] + '.npy'), image_array)    
    
```
